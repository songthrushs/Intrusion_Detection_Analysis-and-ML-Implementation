```python
df['HourTime'] = pd.to_datetime(df['HourTime'], format='%H:%M:%S')
df['Hour'] = df['HourTime'].dt.hour
df['Minute'] = df['HourTime'].dt.minute
df['Second'] = df['HourTime'].dt.second
```

```python
# host_unique = df['HostName_1'].unique()
# ip_counts = df['HostName_1'].value_counts()
# filtered_ips = ip_counts[ip_counts < 60].index

for i in range(6,10):
    cols_plot = ['Hour', 'Minute', 'Second']
    axes = df[df['HostName_1'] == filtered_ips[i]][cols_plot].plot(marker='.', alpha=1, linestyle='None', figsize=(22, 18), subplots=True)
    for ax in axes:
        ax.set_ylabel('Daily Totals') 
        ax.set_title(filtered_ips[i], fontsize=16)
    
```
